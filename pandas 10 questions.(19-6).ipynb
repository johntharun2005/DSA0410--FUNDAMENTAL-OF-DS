```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv("C:/R PROGEAMMING/marks.csv")

# i) Display the first 5 rows
print("First 5 rows:")
print(df.head())

# ii) Display the last 5 rows
print("\nLast 5 rows:")
print(df.tail())

# iii) Purpose of describe() function
print("\nDescription of dataset:")
print(df.describe())
# describe() gives count, mean, std, min, max, and quartiles for numeric columns.

# iv) Select only 3rd to 6th rows (index 2 to 5)
print("\nRows 3 to 6:")
print(df.iloc[2:6])

# v) Select only rows where M3 > 84
print("\nRows where M3 > 84:")
print(df[df['M3'] > 84])

# vi) Filter rows with missing values
print("\nRows with missing values:")
print(df[df.isnull().any(axis=1)])

# vii) Remove rows with missing values
df_cleaned = df.dropna()
print("\nData after removing missing values:")
print(df_cleaned)

# viii) Sort M1 column in descending order
print("\nData sorted by M1 in descending order:")
print(df.sort_values(by='M1', ascending=False))

# ix) Plot the table (marks comparison)
df.plot(kind='bar', x='NAME', y=['M1', 'M2', 'M3'], figsize=(10,6))
plt.title("Student Marks Comparison")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.grid(True)
plt.tight_layout()
plt.show()

# x) Output of marks.ix[3:6, ['m2','m3']] (ix is deprecated)
print("\nEquivalent of marks.ix[3:6, ['m2','m3']]:")
print(df.loc[3:6, ['M2', 'M3']])
```
