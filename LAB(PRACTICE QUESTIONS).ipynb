```python

import numpy as np

student_scores = np.array([
    [80, 75, 90, 70],
    [85, 78, 88, 72],
    [90, 92, 85, 65],
    [70, 80, 78, 60]
])  # 4x4 dummy data

avg_scores = student_scores.mean(axis=0)
subjects = ['Math', 'Science', 'English', 'History']
highest_subject = subjects[np.argmax(avg_scores)]
print("Average Scores per Subject:", dict(zip(subjects, avg_scores)))
print("Subject with Highest Average Score:", highest_subject)
import numpy as np
sales_data = np.array([
    [120, 150, 100],
    [200, 180, 220],
    [90, 160, 140]
])
avg_price = sales_data.mean()
print("Average Price of All Products:", avg_price)
import numpy as np
house_data = np.array([
    [3, 1500, 300000],
    [5, 2000, 500000],
    [6, 2500, 600000],
    [4, 1800, 400000]
])
filtered = house_data[house_data[:, 0] > 4]
avg_price = filtered[:, 2].mean()
print("Average Sale Price for Houses with >4 Bedrooms:", avg_price)
import numpy as np
sales_data = np.array([10000, 12000, 15000, 20000])
total_sales = sales_data.sum()
percent_increase = ((sales_data[3] - sales_data[0]) / sales_data[0]) * 100
print("Total Yearly Sales:", total_sales)
print("Percentage Increase from Q1 to Q4:", round(percent_increase, 2), "%")
import numpy as np
fuel_efficiency = np.array([25.5, 30.0, 22.3, 35.2, 28.9])
avg_efficiency = fuel_efficiency.mean()
improvement = ((fuel_efficiency[3] - fuel_efficiency[0]) / fuel_efficiency[0]) * 100
print("Average Fuel Efficiency:", round(avg_efficiency, 2))
print("Percentage Improvement from Car 1 to Car 4:", round(improvement, 2), "%")
prices = [100, 200, 150]
quantities = [2, 1, 3]
discount_rate = 10  # in percent
tax_rate = 8  # in percent
subtotal = sum(p * q for p, q in zip(prices, quantities))
discount = subtotal * (discount_rate / 100)
tax = (subtotal - discount) * (tax_rate / 100)
total_cost = subtotal - discount + tax
print("Total Purchase Cost:", round(total_cost, 2))
import pandas as pd

order_data = pd.DataFrame({
    'customer_id': [1, 2, 1, 3, 2, 1],
    'order_date': pd.to_datetime(['2024-06-01', '2024-06-02', '2024-06-03', '2024-06-04', '2024-06-05', '2024-06-06']),
    'product': ['A', 'B', 'A', 'C', 'B', 'C'],
    'quantity': [2, 3, 1, 4, 2, 1]
})

print("Total Orders per Customer:\n", order_data.groupby('customer_id').size())
print("Average Quantity per Product:\n", order_data.groupby('product')['quantity'].mean())
print("Earliest Date:", order_data['order_date'].min())
print("Latest Date:", order_data['order_date'].max())
df = pd.DataFrame({
    'product': ['A', 'B', 'C', 'A', 'B', 'A', 'C', 'C', 'B', 'A'],
    'quantity': [2, 1, 3, 2, 2, 1, 4, 3, 2, 1]
})

top_products = df.groupby('product')['quantity'].sum().sort_values(ascending=False).head(5)
print("Top 5 Most Sold Products:\n", top_products)
df = pd.DataFrame({
    'property_id': [101, 102, 103, 104, 105],
    'location': ['CityA', 'CityB', 'CityA', 'CityC', 'CityB'],
    'bedrooms': [3, 5, 4, 6, 3],
    'area_sqft': [1500, 2500, 1800, 3000, 1600],
    'price': [300000, 500000, 350000, 600000, 320000]
})

print("Average Price per Location:\n", df.groupby('location')['price'].mean())
print("Properties with >4 Bedrooms:\n", df[df['bedrooms'] > 4].shape[0])
print("Property with Largest Area:\n", df.loc[df['area_sqft'].idxmax()])
import matplotlib.pyplot as plt
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [10000, 12000, 15000, 13000, 16000]
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales Line Plot")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
plt.bar(months, sales, color='skyblue')
plt.title("Monthly Sales Bar Plot")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
