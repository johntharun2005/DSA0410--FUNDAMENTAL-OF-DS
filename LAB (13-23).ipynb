```python
import pandas as pd
from collections import Counter
import string

data = {
    'ReviewID': [1, 2, 3, 4],
    'ReviewText': [
        "Great product, really loved it!",
        "Good quality, but too expensive.",
        "Amazing product, worth the price.",
        "Not bad, but expected better quality."
    ]
}
df = pd.DataFrame(data)

# Combine all reviews, lowercase, and remove punctuation
all_reviews = ' '.join(df['ReviewText'].str.lower())
all_reviews = all_reviews.translate(str.maketrans('', '', string.punctuation))

# Split into words and count
words = all_reviews.split()
word_freq = Counter(words)

# Display word frequency
print("Frequency distribution of words in customer reviews:")
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

data = {
    'age': [23, 23, 27, 27, 39, 41, 47, 49, 50, 52, 54, 54, 56, 57, 58, 58, 60, 61],
    '%fat': [9.5, 26.5, 7.8, 17.8, 31.4, 25.9, 27.4, 27.2, 31.2,
             34.6, 42.5, 28.8, 33.4, 30.2, 34.1, 32.9, 41.2, 35.7]
}
df = pd.DataFrame(data)
stats = df.agg(['mean', 'median', 'std'])

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
df.boxplot(column='age')
plt.title('Age Distribution')

plt.subplot(2, 2, 2)
df.boxplot(column='%fat')
plt.title('Body Fat Percentage')

plt.subplot(2, 2, 3)
plt.scatter(df['age'], df['%fat'], c='teal', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('% Fat')
plt.grid(True, linestyle='--', alpha=0.7)

plt.subplot(2, 2, 4)
sm.qqplot(df['age'], line='s')
sm.qqplot(df['%fat'], line='s')
plt.tight_layout()
plt.show()

print("Statistical Summary:\n", stats.round(2))
```

```python
import pandas as pd
import string
from collections import Counter
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')

# Sample data as list, not CSV
data = {
    'feedback': [
        "The product is excellent and the delivery was fast!",
        "I love the design and quality of the item.",
        "Very poor customer service and slow delivery.",
        "The product is not worth the money, very disappointed.",
        "Amazing quality, fast shipping, great support!",
        "Product arrived late and damaged. Bad experience."
    ]
}

df = pd.DataFrame(data)
stop_words = set(stopwords.words('english'))

df['feedback'] = df['feedback'].str.lower().apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))

def preprocess_text(text):
    words = text.split()
    return [word for word in words if word not in stop_words]

df['processed_feedback'] = df['feedback'].apply(preprocess_text)
all_words = [word for feedback in df['processed_feedback'] for word in feedback]
word_freq = Counter(all_words)

N = 5  # Top N words
top_n_words = word_freq.most_common(N)

print(f"\nTop {N} most frequent words:")
for word, freq in top_n_words:
    print(f"{word}: {freq}")

# Plot
words, frequencies = zip(*top_n_words)
plt.figure(figsize=(10, 6))
plt.bar(words, frequencies, color='skyblue')
plt.title(f"Top {N} Most Frequent Words in Customer Feedback")
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

data = {
    'Date': ['2025-06-01', '2025-06-01', '2025-06-02', '2025-06-03'],
    'Product': ['Laptop', 'Smartphone', 'Tablet', 'Smartphone'],
    'Quantity Sold': [3, 5, 2, 4],
    'Unit Price': [50000, 20000, 15000, 20000]
}

df = pd.DataFrame(data)
df['Total Sales'] = df['Quantity Sold'] * df['Unit Price']
product_sales = df.groupby('Product').agg({'Total Sales': 'sum'}).reset_index()
product_sales['Profit'] = product_sales['Total Sales'] * 0.20
overall_profit = product_sales['Profit'].sum()
top_products = product_sales.nlargest(5, 'Profit')

print("Total Sales per Product:")
print(product_sales.sort_values('Total Sales', ascending=False))
print(f"\nOverall Company Profit: ₹{overall_profit:,.2f}")
print("\nTop Profitable Products:")
print(top_products[['Product', 'Profit']])
```

```python
import pandas as pd

data = {
    'Customer ID': [1, 2, 3, 4, 5, 6],
    'Age': [25, 40, 35, 28, 50, 31],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female'],
    'Total Spending': [1500, 5000, 3000, 1000, 7000, 2500]
}

df = pd.DataFrame(data)
quantiles = df['Total Spending'].quantile([0.33, 0.67])

df['Spending Segment'] = pd.cut(df['Total Spending'],
                                bins=[-1, quantiles[0.33], quantiles[0.67], float('inf')],
                                labels=['Low Spenders', 'Medium Spenders', 'High Spenders'])

avg_age = df.groupby('Spending Segment')['Age'].mean()

print("Customer Segmentation:")
print(df[['Customer ID', 'Spending Segment']])
print("\nAverage Age per Spending Segment:")
print(avg_age)
print("\nGender Distribution:")
print(df['Gender'].value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': ['2025-01-01', '2025-01-02', '2025-01-03', '2025-02-01', '2025-02-02', '2025-03-01'],
    'Temperature (Celsius)': [22.5, 23.0, 21.5, 24.0, 24.5, 26.0]
}

df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')

monthly_avg_temp = df.groupby('Month')['Temperature (Celsius)'].mean().reset_index()
monthly_avg_temp['Month'] = monthly_avg_temp['Month'].dt.to_timestamp()

plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Temperature (Celsius)'], label='Daily Temp')
plt.plot(monthly_avg_temp['Month'], monthly_avg_temp['Temperature (Celsius)'], color='red', label='Monthly Avg')
plt.title('Temperature Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

orders_data = {
    'Order ID': [1, 2, 3, 4],
    'Customer ID': [101, 102, 101, 102],
    'Order Date': ['2025-01-01', '2025-01-02', '2025-01-10', '2025-01-20']
}

customer_info = {
    'Customer ID': [101, 102],
    'Name': ['Alice', 'Bob'],
    'Email': ['alice@example.com', 'bob@example.com'],
    'Phone Number': ['1234567890', '0987654321']
}

orders_df = pd.DataFrame(orders_data)
customers_df = pd.DataFrame(customer_info)

merged_df = pd.merge(orders_df, customers_df, on="Customer ID", how="inner")
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'])
merged_df = merged_df.sort_values(by=['Customer ID', 'Order Date'])
merged_df['Time Diff'] = merged_df.groupby('Customer ID')['Order Date'].diff()

print(merged_df)
print("\nAverage time between consecutive orders:")
print(merged_df['Time Diff'].dropna().mean())
```
